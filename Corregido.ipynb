VIP: Para ver los tipos de cada columna

```python
def ver_tipos_columnas(df,lista_columnas):
    for x in lista_columnas:
        print(df[x].apply(type).value_counts())
        print("---------------------------")

print(ver_tipos_columnas(df,lista_columnas))

```

```python
def eliminar_minoritarios(df, lista_columnas):
    for col in lista_columnas:
        print(f"\nProcesando columna: {col}")

        # Detectar tipo mayoritario
        tipo_mayoritario = df[col].apply(type).value_counts().idxmax()
        print(f"Tipo mayoritario en {col}: {tipo_mayoritario}")

        # Filtrar solo filas del tipo mayoritario
        df = df[df[col].apply(lambda v: isinstance(v, tipo_mayoritario))]

    return df

eliminar_minoritarios(df,lista_columnas)
print(ver_tipos_columnas(df,lista_columnas))
```

Creo mis listas de limpieza

```python
lista_strings_a_Capital = ["Type","Country","Activity","Sex","Species","FatalY/N"]
lista_string_numeros = ["Age"]
lista_float_int = ['Year']
```

Creo mis funciones de limpieza


```python
def strings_a_Capital(df, lista_columnas):
    for col in lista_columnas:
        df[col] = (
            df[col]
            .where(df[col].notna())   # mantiene NaN
            .astype(str)
            .str.strip()
            .str.title()
        )
    return df


```

```python
def convertir_edad(df, columnas):
    for col in columnas:
        # 1) Convertir a string para limpiar
        df[col] = df[col].astype("string").str.strip()

        # 2) Reemplazar valores basura
        basura = ["nan", "none", "unknown", "na", "", " ", "null", "n/a"]
        df[col] = df[col].replace(basura, pd.NA)

        # 3) Convertir a número
        df[col] = pd.to_numeric(df[col], errors="coerce")

        # 4) Convertir floats tipo 25.0 a enteros
        df[col] = df[col].astype("Int64")

        # 5) Eliminar edades imposibles
        df.loc[(df[col] < 1) | (df[col] > 120), col] = pd.NA

    return df



```

6.3. Categoría simplificada de actividad

```python


def categorize_activity(x):
    if "surf" in x:
        return "Surf"
    if "swim" in x or "bathing" in x:
        return "Natación"
    if "fish" in x:
        return "Pesca"
    if "dive" in x:
        return "Buceo"
    if "kayak" in x or "canoe" in x:
        return "Kayak"
    return "Otras"


df["ActivityCategory"] = df["Activity"].apply(categorize_activity)
```
