6.2. Grupo de edad

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")


df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 30, 45, 60, 120],
    labels=["Niño", "Adolescente", "Joven", "Adulto", "Maduro", "Senior"])
```

6.3. Categoría simplificada de actividad

```python


def categorize_activity(x):
    if "surf" in x:
        return "Surf"
    if "swim" in x or "bathing" in x:
        return "Natación"
    if "fish" in x:
        return "Pesca"
    if "dive" in x:
        return "Buceo"
    if "kayak" in x or "canoe" in x:
        return "Kayak"
    return "Otras"


df["ActivityCategory"] = df["Activity"].apply(categorize_activity)
```
